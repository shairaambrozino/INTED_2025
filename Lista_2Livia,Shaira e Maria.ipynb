```python
#23 Um módulo de análise precisa classificar uma lista de valores. Leia 10 números e exiba quantos são positivos, negativos e quantos são iguais a zero.
pos = neg = zero = 0

for i in range(10):
    num = float(input(“Digite um número: “))
    if num > 0:
        pos += 1
    elif num < 0:
        neg += 1
    else:
        zero += 1

print(f”Positivos: {pos}, Negativos: {neg}, Zeros: {zero}”)
```

```python
#28 Desenvolva um programa que leia 10 números e some apenas os valores positivos. Caso um número negativo seja digitado, utilize continue para ignorá-lo.
soma = 0

for i in range(10):
    num = float(input(“Digite um número: “))
    if num < 0:
        continue
    soma += num

print(f”Soma dos positivos: {soma}”)
```

```python
#29 Simule um acumulador que percorre números até que a soma ultrapasse 200. Quando isso ocorrer, encerre o laço usando break.
soma = 0
num = 1

while True:
    soma += num
    num += 1
    if soma > 200:
        break

print(f”A soma final foi: {soma}”)
```

```python
#30  Um programa de edição quer contar quantas palavras o usuário digitou. Continue lendo palavras até que o usuário digite “fim”. Depois, exiba quantas palavras válidas foram informadas.
cont = 0

while True:
    palavra = input(“Digite uma palavra (‘fim’ para encerrar): “).lower()
    if palavra == “fim”:
        break
    cont += 1

print(f”Você digitou {cont} palavras.”)
```
